```python
import requests
import re
from bs4 import BeautifulSoup
import json
import pandas as pd
domain_name = 'https://www.dice.com'
```

```python
## Choose you want to select "Work_type","Location"
Key_work = 'Front End Developer'
Key_location = 'New York, NY'
## Set homepage
key_main='https://www.dice.com/jobs/q-'+Key_work+'-l-'+Key_location+'-radius-30-startPage-0-jobs?'
```

```python
## Calculate pages
response_main = requests.get(key_main)
soup = BeautifulSoup(response_main.text,'lxml')
pages = int(str(soup.find("",{"id":"posiCountId"}).text).replace(',',''))//30
```

```python
## Add all pages's work
out_page = []
herf = []
for i in range(pages):
    response_page = requests.get('https://www.dice.com/jobs/q-'+Key_work+'-l-'+Key_location+'-radius-30-startPage-'+str(i+1)+'-jobs?')
    soup = BeautifulSoup(response_page.text,'lxml')
    # page contains 30 urls
    for i in range(30):
        herf.append(domain_name + soup.find("",{"id":"position"+str(i)}).get('href'))
```

```python
## Crawler Data
for i,work in enumerate(herf):
    ## Beause some works are broken or missing, I use "try except" to avoid error.
    try:
        response_work = requests.get(work)
        soup = BeautifulSoup(response_work.text,'lxml')
        ## Crawler we need tags
        jobTitle = soup.find("",{"class":"jobTitle"}).text
        ## replace some extra symbol
        jobEmployer = soup.find("",{"class":"employer"}).text.replace('\n','').replace('\t','')
        jobLocation = soup.find("",{"class":"location"}).text.replace('\n','')
        jobPostTime = soup.find("",{"class":"posted hidden-xs"}).text
        out = []
        ## tag job-info is a array .
        foo = soup.find_all("",{"class":"row job-info"})
        for o in foo:
            out.append(o.text.replace('\n','').replace('\t',''))
            
    except AttributeError:
        jobTitle='None'
        jobEmployer='None'
        jobLocation='None'
        jobPostTime='None'
        foo='None'
    
        
    try:
        ## Beause the work of dice.com's html format is not fixed,I use "try except" to avoid error.
        skills = out[0]
        employmentType = out[1]
        baseSalary = out[2]   
    except IndexError:
        skills = 'None'
        employmentType = 'None'
        baseSalary = 'None'
        
    ## Jobdescription
    str1 = str(soup.find("",{"id":"jobdescSec"}))
    soup = BeautifulSoup(str1.replace('<br/>','\n'),'lxml')
    jobDescription = soup.text
    ## work's url
    url = work ##
    
    ## Save data to json format 
    if i ==0:
        data = {"Front-End-Developer":[{"jobTitle":jobTitle,
                 "jobEmployer":jobEmployer,
                 "jobLocation":jobLocation,
                 "jobPostTime":jobPostTime,
                 "skills":skills,
                 "employmentType":employmentType,
                 "baseSalary":baseSalary,
                 "jobDescription":soup.text,
                 "url":url}]}
    else:
        add_data = {     "jobTitle":jobTitle,
                         "jobEmployer":jobEmployer,
                         "jobLocation":jobLocation,
                         "jobPostTime":jobPostTime,
                         "skills":skills,
                         "employmentType":employmentType,
                         "baseSalary":baseSalary,
                         "jobDescription":soup.text,
                         "url":url                          }
        data['Front-End-Developer'].append(add_data)
    if i%10==0:
        print('deal with'+ str(i) +' work is OK')
```

```python
## save json data to XXXX.json
with open("Front-End-Developer-NY-3000.json","w") as f:
    json.dump(data,f)
    print('ok')
```
